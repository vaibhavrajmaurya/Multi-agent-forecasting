```python
# Write a Pandas function to find the correlation between the first two columns.
correlation = df_iris2d_clean['sepal_length'].corr(df_iris2d_clean['sepal_width'])
```

```python
# Write a Pandas function to find the mean, median and standard deviation of column 3.
mean_col3 = df_iris2d_clean['petal_length'].mean()
median_col3 = df_iris2d_clean['petal_length'].median()
std_col3 = df_iris2d_clean['petal_length'].std()
```
